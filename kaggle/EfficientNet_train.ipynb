```python
print("copying codes: ", CODES_PATH, ".")
import shutil
for dirpath, dirnames, filenames in os.walk(CODES_PATH):
    for filename in filenames:
        src_path = os.path.join(dirpath, filename)
        print(src_path)
        dst_path = dirpath.replace(SUPPLEMENT_INPUT_PATH, "")
        dst_path = os.path.join(dst_path, filename)
        dst_path = os.path.join(".", dst_path)
        print(dst_path)
        if not os.path.isdir(dst_path):
            print("not exist")
            os.makedirs(dst_path, exist_ok=False)
        shutil.copy(src_path, dst_path)
print("================================")
import sys
sys.path.append("/efficientnet_codes")
```
